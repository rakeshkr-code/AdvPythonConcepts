### Threads

```python
import threading
import time
```

```python
def crawl(link, delay=3):
    print(f"crawl started for <{link}>")
    time.sleep(delay)  # Blocking I/O (simulating a network request)
    print(f"crawl ended for <{link}>")
```

```python
links = [
    # "https://python.org",
    # "https://docs.python.org",
    # "https://peps.python.org",
    "link-1",
    "link-2",
    "link-3",
    "link-4",
    "link-5",
    "link-6",
    "link-7",
    "link-8",
    "link-9",
    "link-10",
]
# Start threads for each link
threads = []
for link in links:
    # Using `args` to pass positional arguments and `kwargs` for keyword arguments
    t = threading.Thread(target=crawl, args=(link,), kwargs={"delay": 10})
    threads.append(t)
print(f"Links: {links}\nThreads: {threads}")
```

```python
# Start each thread
for t in threads:
    t.start()
    print(f"Started thread [{t.name}] for link <{links[threads.index(t)]}>")
```

```python
# Wait for all threads to finish
for t in threads:
    t.join()
    print(f"Thread {t.name} finished")
```

**Obs:-** 
- 10 different processes / functions or tasks started at 10 different threads one by one
- Threads didn't completel sequentially, ie, some thread (function) started earlier may be late to finish, and some thread (function) started late may end up early
- But during thread join it joins threads sequentially as it started, or it waits for the threads that takes time to complete, after all threads (functions) complete, then it joins one by one.
